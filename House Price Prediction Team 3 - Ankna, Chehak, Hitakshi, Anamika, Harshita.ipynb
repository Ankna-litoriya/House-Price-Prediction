```python
#importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data Processing**

```python
#extracting the training data
data = pd.read_csv("house pricing.csv")
```

```python
#show the extracted data
data
```

```python
#show he info in columns
data.info()
```

```python
#removes the null values and saves the table as original table
data.dropna(inplace=True)
```

```python
#all the non null values
data.info()
```

 **Data Splitting**

```python
#split the data
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

```python
#training data and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)
```

```python
#joining the whole training data
train_data = x_train.join(y_train)
```

```python
#display data
train_data
```

```python
#show data in histogram
train_data.hist(figsize=(15,10))
```

Heatmap distribution

```python
#display data correlation in heatmap
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap='YlGnBu')
```

```python
#taking log of values that are one sided in histogram
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
#display the histograms
train_data.hist(figsize=(15,8))
```

```python
#values in ocean proximity
train_data.ocean_proximity.value_counts()
```

```python
#convert values in ocean proximity to numeric
#hot encoding
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)
```

```python
train_data
```

```python
#display data correlation in heatmap
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap='YlGnBu')
```

```python
#heatmap for relation btn latitude and longitude
plt.figure(figsize=(15, 8))
sns.scatterplot(x="latitude", y="longitude", data= train_data, hue="median_house_value", palette="magma")
```

```python
#correlating different values
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_ratio'] = train_data['total_rooms']/train_data['households']
```

```python
#display data correlation in heatmap
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap='YlGnBu')
```

Linear Regression

```python
#linear regression
from sklearn.linear_model import LinearRegression
x_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
#making test data accoring to training data
test_data = x_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

test_data['bedroom_ratio'] = test_data['total_rooms']/test_data['total_bedrooms']
test_data['household_ratio'] = test_data['total_rooms']/test_data['households']
```

```python
#separating the value of house from rest data
x_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
#final test data
test_data
```

```python
#final training data
train_data
```

```python
reg.score(x_test, y_test)
```

Random Forest Regression

```python
#forest regressor
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

```python
#Grid search cross validation

from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [30, 50, 100],
    "max_features": [8, 10, 12, 15]
}
grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring = "neg_mean_squared_error",
                           return_train_score = True)
grid_search.fit(x_train, y_train)
```

```python
best_forest = grid_search.best_estimator_
```
